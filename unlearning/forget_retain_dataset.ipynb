Cretes Forget and Retain dataset

```python
import json
from datasets import load_dataset
import random
import math
```

## Forget

```python
forget_entity = "Elon Musk"

with open(f"metadata/{forget_entity.replace(' ', '_')}.json", 'r') as fp:
    data = json.load(fp)

qa_data = data["qa_pairs"]

random.shuffle(qa_data)

split_index = math.ceil(len(qa_data) * 0.7)

train_set = qa_data[:split_index]
test_set = qa_data[split_index:]

print(f"Train size: {len(train_set)}")
print(f"Test size: {len(test_set)}")

with open(f"dataset/train_forget_set_{forget_entity.replace(' ', '_')}.json", 'w') as fp:
    json.dump(train_set, fp, indent=4)

with open(f"dataset/test_forget_set_{forget_entity.replace(' ', '_')}.json", 'w') as fp:
    json.dump(test_set, fp, indent=4)
```

## Retain set

```python
entities = ["Peter Thiel", "Bill Gates", "Sam Altman", "Jeff Bezos", "Mark Zuckerberg"]

entities_data= {}

for entity in entities:
    with open(f"metadata/{entity.replace(' ', '_')}.json", 'r') as fp:
        entities_data[entity] = json.load(fp)

print({entity: len(entity_data["qa_pairs"]) for entity, entity_data in entities_data.items()})
```

```python
train = []
test = []

for entity in entities:
    data = entities_data[entity]
    
    qa_pairs = data["qa_pairs"]
    random.shuffle(qa_pairs)

    split_index = min(math.ceil(len(qa_pairs) * 0.7), 70)

    entity_train = qa_pairs[:split_index]
    entity_test = qa_pairs[split_index: split_index + 30]

    train.extend(entity_train)
    test.extend(entity_test)

    print(entity, len(entity_train), len(entity_test))

random.shuffle(train)
random.shuffle(test)

with open(f"dataset/train_retain_set_{forget_entity.replace(' ', '_')}.json", 'w') as fp:
    json.dump(train, fp, indent=4)

with open(f"dataset/test_retain_set_{forget_entity.replace(' ', '_')}.json", 'w') as fp:
    json.dump(test, fp, indent=4)
```
